## This notebook counts the number of TRUE claims for each speaker

It takes an excel file with the combined LIAR set (LIAR.xls), so that the value for the column occurs on all the set and conforms to how the values for other scores are calculated.

It outputs the truth column in a separate Excel file which makes it possible to add the column manually.
train, valid and test sets are then separated manually.

### Important
##### For this code to give the desired result the entire LIAR set has to be sorted By Speaker and then by ID in ascending order.

```python
import pandas as pd
LIAR=pd.read_excel('LIAR.xls')
scores={True: 5, 'mostly-true': 4, 'half-true': 3, 'barely-true': 2, False: 1, 'pants-fire': 0}
LIAR.Label = [scores[item] for item in LIAR.Label]
```

```python

lastRowSpeaker=' '
truths=0


maxCounts=[] #will contain the maximum number of statements said by the Speaker

for row in range (len(LIAR)):
    ThisSpeaker=LIAR.iloc[row,4]
    
    if ThisSpeaker!=lastRowSpeaker: 
        if row!=0: maxCounts.append(truths)
        truths=0 #restart couting of truths from zero
        
            
    if LIAR.iloc[row,1]==5:
        truths=truths+1
        

    lastRowSpeaker=ThisSpeaker

maxCounts.append(truths)
 
```

```python

lastRowSpeaker=' '
 
countsRow=-1
count=[]

 
for row in range (len(LIAR)):
    ThisSpeaker=LIAR.iloc[row,4]
    
    if ThisSpeaker!=lastRowSpeaker: 
        countsRow+=1
        
    count.append(maxCounts[countsRow])
    lastRowSpeaker=ThisSpeaker

    
LIAR['AllTruths']=count
LIAR.tail(30)
```

```python
for row in range (len(LIAR)):
    print(LIAR.iloc[row,4],LIAR.iloc[row,1], count[row])
```

```python
pd=LIAR['AllTruths']

pd.to_excel('TotalTruthsColumn.xls' , index=False)
```
