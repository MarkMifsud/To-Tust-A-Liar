This notebook creates incremental reputations scores

The reputation scores do not include the score for the statement in the same row; only scores for past statements in this dataset.  The ID is used to keep the sequence.

To run this code make sure you have all of liar in one file sorted by Speaker followed by ID

```python
import pandas as pd
```

```python
LIAR=pd.read_excel('LIAR-ordered.xlsx' , index=False)
LIAR
```

```python

lastRowSpeaker=' '
truths=0
mostlies=0
halfs=0
barelys=0
fakes=0
pants=0
counts=0

real=[]
most=[]
half=[]
barely=[]
fake=[]
pant=[]
count=[]

maxCounts=[] #will contain the maximum number of statements said by the Speaker

for row in range (len(LIAR)):
    ThisSpeaker=LIAR.iloc[row,4]
    
    if ThisSpeaker!=lastRowSpeaker: 
        truths=0
        mostlies=0
        halfs=0
        barelys=0
        fakes=0
        pants=0
        
        maxCounts.append(counts)
        counts=0
        
    if LIAR.iloc[row,1]==5:
        truths=truths+1

    if LIAR.iloc[row,1]==4:
        mostlies=mostlies+1

    if LIAR.iloc[row,1]==3:
        halfs=halfs+1

    if LIAR.iloc[row,1]==2:
        barelys=barelys+1

    if LIAR.iloc[row,1]==1:
        fakes=fakes+1

    if LIAR.iloc[row,1]==0:
        pants=pants+1
        
    counts=counts+1

    real.append(truths)
    most.append(mostlies)
    half.append(halfs)
    barely.append(barelys)
    fake.append(fakes)
    pant.append(pants)
    
    
    lastRowSpeaker=ThisSpeaker

maxCounts.append(counts)
LIAR['PantsINC']=pant
LIAR['NotRealINC']=fake
LIAR['BarelyINC']=barely
LIAR['HalfINC']=half
LIAR['MostlyINC']=most
LIAR['RealINC']=real

LIAR 
```
