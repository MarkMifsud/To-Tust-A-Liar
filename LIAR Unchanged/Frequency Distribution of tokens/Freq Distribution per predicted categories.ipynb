In this notebook we try to see if the fake-news classification done by BERT correlates with the presence of certain words/terms.

The reason fro this is two fold:

1)  We want to see if BERT correlates veracity to tokens found, as Sentiment Analysis would be.

2)  We can check if for specific Speakers, the use of certain words are indicative of veracity or lack thereof.

```python
import torch
#import torch.nn as nn
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
from transformers import BertTokenizer, BertModel
device

model_class='bert'  # bert or roberta or albert
model_version='bert-base-cased' #bert-base-cased, roberta-base, roberta-large, albert-base-v2 OR albert-large-v2
output_folder='./TunedModels/'+model_class+'/'+model_version+"/"

# Load pre-trained model (weights)
CheckPoint='checkpoint-322-epoch-2'  #epoch 2
preSavedCheckpoint=output_folder+CheckPoint
model =BertModel.from_pretrained(preSavedCheckpoint) # ,output_hidden_states=True  if you wish

# Load pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained(preSavedCheckpoint)

#for testing it (uncomment the following 2 lines)
tokens= tokenizer.tokenize("Hello, my dog's ear is very very cute")
tokens[0]
```

Now we create a dictionary for each class.

In a dictionary we store each token and its count within that class.

```python
''' Now we can  attempt to see if each category correlates with certain words

we first need to form groups of statements, separate by categories.

Then we attempt to inspect if certain words correlate with certain categories more than others
'''

PantsString=''
FakeString=''
MostlyFakeString=''
HalfTruthString=''
MostlyRealString=''
RealString=''

PantsArray=[]
FakeArray=[]
MostlyFakeArray=[]
HalfTruthArray=[]
MostlyRealArray=[]
RealArray=[]

for row in range(len(Pred)):
    
     
    if Pred[row]==0:
        PantsString=PantsString+' '+ReadSet.iloc[row,2]
        PantsArray.append(ReadSet.iloc[row,2])
        
    if Pred[row]==1:
        FakeString=FakeString+' '+ReadSet.iloc[row,2]
        FakeArray.append(ReadSet.iloc[row,2])
        
    if Pred[row]==2:
        MostlyFakeString=MostlyFakeString+' '+ReadSet.iloc[row,2]
        MostlyFakeArray.append(ReadSet.iloc[row,2])
    
    if Pred[row]==3:
        HalfTruthString=HalfTruthString+' '+ReadSet.iloc[row,2] 
        HalfTruthArray.append(ReadSet.iloc[row,2])
    
    if Pred[row]==4:
        MostlyRealString=MostlyRealString+' '+ReadSet.iloc[row,2] 
        MostlyRealArray.append(ReadSet.iloc[row,2])
        
    if Pred[row]==5:
        RealString=RealString+' '+ReadSet.iloc[row,2]
        RealArray.append(ReadSet.iloc[row,2])
        
AllStrings= RealString+' '+MostlyRealString+' '+HalfTruthString+' '+MostlyFakeString+' '+FakeString+' '+PantsString

```

```python
# a funtion that takes a string (concatenation of statements in the same category)
# and returns a pyhton dictionary with the frequnecy distribution for each token
#this one considers even multiple appearances of the token in the same row

def TokenDistribution(textstring):
    myText =tokenizer.tokenize(textstring)
    freqDist = FreqDist(myText)

    words = freqDist.keys()
    #for x in words:
            #print(freqDist[x],end='\t')
            #print (x )
    return freqDist
        
```

```python
# first we create the table with the column of unique tokens

DistribTable=pd.DataFrame(columns=['Tokens','Real', 'MostlyReal', 'HalfTruth','MostlyFake','Fake','Pants'])


AllTokensDistribution=TokenDistribution(AllStrings)

uniqueTokens=[]

for x in AllTokensDistribution:
    uniqueTokens.append(x)
    
DistribTable['Tokens']=uniqueTokens 
#del(AllStrings, AllTokensDistribution )
DistribTable
```

Now that we have the table, we can produce the distribution of tokens.

#### HOWEVER

We need to count how many statements (in each category) have that token, not how many tokens exist in that category.

We need to define a function that takes a statement and ignores duplicate tokens from it.

```python
def RemoveRepeatedTokens(StatementsArray):
    
    dict={}
    
    for row in StatementsArray:
        tokensCount=TokenDistribution(row)
        
        for x  in tokensCount :
            if x in dict:
                dict[x]=dict[x]+1
            else:
                dict[x]=1
 

    for row in uniqueTokens:
        if row not in dict:
            dict[row]=0

        
        
    return dict
```

Now we need to compare the numbers for each token.

According to our hypothesis: Any tokens are are particularly high within their category and low in others can be the cause of the statement being predicted as its respective predicted classification.

To compare we need a Table where the first column will be the tokens, the subsequent columns will be the population of that token for each class.  Once that table is available, comparisons can be made to accept or reject our hypothesis.

```python
# now we create a row of integers for each category
# each integer will be the distribution for that token in that category


RealDistribution= RemoveRepeatedTokens(RealArray)
MostlyRealDistribution =RemoveRepeatedTokens(MostlyRealArray)
HalfTruthDistribution =RemoveRepeatedTokens(HalfTruthArray)
MostlyFakeDistribution =RemoveRepeatedTokens(MostlyFakeArray)
FakeDistribution =RemoveRepeatedTokens(FakeArray)
PantsDistribution =RemoveRepeatedTokens(PantsArray)

#del(RealString,MostlyRealString,HalfTruthString,MostlyFakeString,FakeString,PantsString)

for row in range(len(DistribTable)):
    currentToken=DistribTable.iloc[row,0]
    DistribTable.iloc[row,1]=RealDistribution[currentToken]
    DistribTable.iloc[row,2]=MostlyRealDistribution[currentToken]
    DistribTable.iloc[row,3]=HalfTruthDistribution[currentToken]
    
     
    DistribTable.iloc[row,4]=MostlyFakeDistribution[currentToken]
    
    DistribTable.iloc[row,5]=FakeDistribution[currentToken]
    DistribTable.iloc[row,6]=PantsDistribution[currentToken]
   
```
